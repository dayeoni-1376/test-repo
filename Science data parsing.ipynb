```python
from bs4 import BeautifulSoup
import urllib.request as req
```

```python
url2 = "https://www.sciencemag.org/news/latest-news?_ga=2.132947342.1852266082.1583048945-415357262.1580001306&page=%s" 

newsLink2=[] # 뉴스 리스트 추출 W(2019~2020 약 1년 기간 동안)
for i in list(range(1,67)): 
    url_list = url2 % (i)
    html = req.urlopen(url_list)
    bsObject = BeautifulSoup(html,"html.parser")
    link = bsObject.find_all("h2", class_="media__headline") 
    for res in link:
        link_list= res.find("a")
        newsLink2.append(link_list.get("href"))
        
news_url2 = "https://www.sciencemag.org/"

Science=[]
for i in range(len(newsLink2)): 
    html=req.urlopen(news_url2+newsLink2[i])
    bsObject = BeautifulSoup(html,"html.parser")
    content = bsObject.find_all("p")
    for res in content:
        Science.append(res.get_text())
```

```python
import pandas as pd
Sciencedata=pd.DataFrame(Science,columns=['text'])
Sciencedata.to_csv("Science.csv")
```

```python
import re
p= re.compile("Vol |By [A-Za-z]+ [A-Za-z]+. [0-9],")

drop_index=[]
for i in range(len(Sciencedata)):
    if p.findall(str(Sciencedata.text[i]))!=[]:
        print(Sciencedata.text[i])
        drop_index.append(i)
    
```

```python
len(drop_index)
```

```python
Sciencedata=Sciencedata.drop(index=drop_index).reset_index(drop=True) # drop함 
Sciencedata
```
