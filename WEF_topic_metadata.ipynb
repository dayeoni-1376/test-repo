# 데이터 웹 크롤링

```python
from bs4 import BeautifulSoup
import urllib.request as req # url 호출 라이브러리 
```

```python
url = "https://www.weforum.org/reports?page=%s"

rpHeadlines=[] # 제목 리스트 
for i in list(range(1,5)):
    url_list = url % (i)
    html = req.urlopen(url_list)
    bsObject = BeautifulSoup(html,"html.parser")
    res_list = bsObject.find_all("div", class_="tout__details")
    for link in res_list: 
        title_list = link.find("h3", class_="tout__title").get_text()
        rpHeadlines.append(res_list)
```
